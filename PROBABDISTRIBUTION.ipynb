```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
mu, sigma = 0, 0.1
s = np.random.normal(mu, sigma, size = 1000)
count, bins, ignored = plt.hist(s, 30, density = True)
plt.title('Normal Distribution')
plt.show()
```

```python
s = np.random.uniform(-1, 0, 1000)

count, bins, ignored = plt.hist(s, 15, density = True)
plt.title('Uniform Distribution')
plt.show()
```

```python
#3. Cauchy Distribution

s = np.random.standard_cauchy(10000000)
s = s[(s >- 25) & (s<25)]
plt.hist(s, bins = 100, density = True)
plt.title('Cauchy Distribution')
plt.show()
```

#### 4. t Distribution


```python
s = np.random.standard_t(df = 1, size = 100000)
s = s[np.where(abs(s) < 5)]
h = plt.hist(s, bins = 100, density = True)
plt.title('t Distribution')
plt.show()
```

####  5. F distribution

```python
s = np.random.f(dfnum = 1, dfden = 1, size = 10000)
s = s[np.where(abs(s) < 5)]
h = plt.hist(s, bins = 100, density = True)
plt.title('F Distribution')
plt.show()
```

#### 6. chi square Distribution

```python
s = np.random.chisquare(df = 1, size = 10000)
s = s[np.where(abs(s) < 20)]
h = plt.hist(s, bins = 100, density = True)
plt.title('chi square Distribution')
plt.show()
```

#### 7. Exponential distribution

```python
s = np.random.exponential(scale = 1.0, size = 10000)
h = plt.hist(s, bins = 1000, density = True)
plt.title('Exponential Distribution')
plt.show()
```

#### 8. Weibull Distribution

```python
s = np.random.weibull(a = 0.5, size = 10000)
s = s[np.where(abs(s) < 5)]
h = plt.hist(s, bins = 100, density = True)
plt.title('Weibull Distribution')
plt.show()
```

#### 9. Log Normal Distribution

```python
s = np.random.lognormal(mean = 0.0, sigma = 0.5, size = 10000)
s = s[np.where(abs(s) < 5)]
h = plt.hist(s, bins = 100, density = True)
plt.title('Log Normal Distribution')
plt.show()
```

#### 10. Gamma Distribution

```python
s = np.random.gamma(shape = 0.5, scale = 1.0, size = 10000)
s = s[np.where(abs(s) < 10)]
h = plt.hist(s, bins = 100, density = True)
plt.title('Gamma Distribution')
plt.show()
```
